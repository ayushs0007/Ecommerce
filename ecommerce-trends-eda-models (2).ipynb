```python
features = df[['Price', 'Rating', 'NumReviews', 'StockQuantity', 'Discount']]
target = df['Sales']
```

```python
# Splitting the dataset into training and testing sets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
y_pred = lr_model.predict(X_test)
print('Linear Regression RMSE:', mean_squared_error(y_test, y_pred, squared=False))
print('Linear Regression R^2:', r2_score(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import LabelEncoder, StandardScaler
scaler=StandardScaler()
X=features
y=target

# For the sake of this example, let's create a binary target variable
y_binary = (y > y.median()).astype(int)  # 1 if price is above median, 0 otherwise

X_train_binary, X_test_binary, y_train_binary, y_test_binary = train_test_split(X, y_binary, test_size=0.2, random_state=42)
X_train_binary_scaled = scaler.fit_transform(X_train_binary)
X_test_binary_scaled = scaler.transform(X_test_binary)

logreg_model = LogisticRegression(random_state=42)
logreg_model.fit(X_train_binary_scaled, y_train_binary)
y_pred_binary = logreg_model.predict(X_test_binary_scaled)

print('Logistic Regression Accuracy:', accuracy_score(y_test_binary, y_pred_binary))
print('Confusion Matrix:\n', confusion_matrix(y_test_binary, y_pred_binary))
print('Classification Report:\n', classification_report(y_test_binary, y_pred_binary))
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)
y_pred = dt_model.predict(X_test)
print('Decision Tree RMSE:', mean_squared_error(y_test, y_pred, squared=False))
print('Decision Tree R^2:', r2_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)
print('Random Forest RMSE:', mean_squared_error(y_test, y_pred, squared=False))
print('Random Forest R^2:', r2_score(y_test, y_pred))
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gb_model = GradientBoostingRegressor(random_state=42)
gb_model.fit(X_train, y_train)
y_pred = gb_model.predict(X_test)
print('Gradient Boosting RMSE:', mean_squared_error(y_test, y_pred, squared=False))
print('Gradient Boosting R^2:', r2_score(y_test, y_pred))
```

```python
from sklearn.svm import SVR
svr_model = SVR()
svr_model.fit(X_train, y_train)
y_pred = svr_model.predict(X_test)
print('Support Vector Regressor RMSE:', mean_squared_error(y_test, y_pred, squared=False))
print('Support Vector Regressor R^2:', r2_score(y_test, y_pred))
```

```python
# Initialize dictionaries to store scores
rmse_scores = {}
r2_scores = {}
rmse_scores['LR'] = mean_squared_error(y_test, y_pred, squared=False)
r2_scores['LR'] = r2_score(y_test, y_pred)
rmse_scores['DT'] = mean_squared_error(y_test, y_pred, squared=False)
r2_scores['DT'] = r2_score(y_test, y_pred)
rmse_scores['RF'] = mean_squared_error(y_test, y_pred, squared=False)
r2_scores['RF'] = r2_score(y_test, y_pred)
rmse_scores['GB'] = mean_squared_error(y_test, y_pred, squared=False)
r2_scores['GB'] = r2_score(y_test, y_pred)
rmse_scores['SVR'] = mean_squared_error(y_test, y_pred, squared=False)
r2_scores['SVR'] = r2_score(y_test, y_pred)
logreg_accuracy = accuracy_score(y_test_binary, y_pred_binary)
```

```python
# Visualization
models = ['LR', 'DT', 'RF', 'GB', 'SVR']
rmse_values = [rmse_scores[model] for model in models]
r2_values = [r2_scores[model] for model in models]
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(models, rmse_values, color='skyblue')
plt.ylabel('RMSE')
plt.title('RMSE of Different Models')
plt.subplot(1, 2, 2)
plt.bar(models, r2_values, color='salmon')
plt.ylabel('R^2 Score')
plt.title('R^2 Score of Different Models')
plt.tight_layout()
plt.show()
```

```python
# Logistic Regression Accuracy
plt.figure(figsize=(6, 4))
plt.bar(['Logistic Regression'], [logreg_accuracy], color='green')
plt.ylabel('Accuracy')
plt.title('Accuracy of Logistic Regression')
plt.show()
```

```python
results_df = pd.DataFrame({
    'Model': ['Linear Regression', 'Decision Tree', 'Random Forest', 'Gradient Boosting', 'SVR', 'Logistic Regression'],
    'RMSE': [rmse_scores['LR'], rmse_scores['DT'], rmse_scores['RF'], rmse_scores['GB'], rmse_scores['SVR'], None],
    'R^2 Score': [r2_scores['LR'], r2_scores['DT'], r2_scores['RF'], r2_scores['GB'], r2_scores['SVR'], None],
    'Accuracy': [None, None, None, None, None, logreg_accuracy]
})
results_df.to_csv('/kaggle/working/model_results.csv', index=False)
```
